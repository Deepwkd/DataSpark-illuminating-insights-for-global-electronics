```python
#Importing Libraries
import pandas as pd
import numpy as np
```

```python
#Importing all csv
ct= pd.read_csv('Customers.csv', encoding='ISO-8859-1')
er= pd.read_csv('Exchange_Rates.csv',encoding='ISO-8859-1')
pt= pd.read_csv('Products.csv',encoding='ISO-8859-1')
sales= pd.read_csv('Sales.csv',encoding='ISO-8859-1')
stores= pd.read_csv('Stores.csv',encoding='ISO-8859-1')
```

```python
#Data cleaning (Customers.csv)

ct.head(10)
```

```python
ct.rename(columns={
    'State Code': 'StateCode',
    'Zip Code': 'ZipCode',
},inplace=True)
```

```python
ct.info()
```

```python
# Filling null values
ct['StateCode'].fillna('NP', inplace=True)
```

```python
#Filled "NP" State code for Napoli
ct[ct['State']=="Napoli"]
```

```python
#Converting data type

ct['Birthday'] = pd.to_datetime(ct['Birthday'])
```

```python
# Dealing with capitalized values
ct['City'] = ct['City'].str.lower()


```

```python
#Saving the cleaned data 
ct.to_csv('Customers_cleaned.csv', index=False)

```

```python
#Cleaning Exchange Rates

er.head(10)
```

```python
er.shape
```

```python
# Converting data type
er['Date']= pd.to_datetime(er['Date'])
```

```python
#cleaned data
er.to_csv('Exchange_Rates_cleaned.csv', index=False)
```

```python
#Cleaning Product data

pt.head(10)
```

```python
pt.rename(columns={
    'Product Name': 'ProductName',
    'Unit Cost USD': 'UnitCostUSD',
    'Unit Price USD': 'UnitPriceUSD',
},inplace=True)
```

```python
pt.info()
```

```python
# Finding rows where 'Unit Cost USD' contains 'comma'
pt[pt['UnitCostUSD'].str.contains(',')]


```

```python
#Removing commas , dollar symbol and converting into float
pt['UnitCostUSD']=pt['UnitCostUSD'].str.replace('$','').str.replace(',', '')
pt['UnitCostUSD'] = pt['UnitCostUSD'].astype(float)

pt['UnitPriceUSD']=pt['UnitPriceUSD'].str.replace('$','').str.replace(',','')
pt['UnitPriceUSD'] = pt['UnitPriceUSD'].astype(float)
```

```python
#saving cleaned data
pt.to_csv('Products_cleaned.csv', index=False)
```

```python
#Cleaning sales data

sales.head(20)
```

```python
sales.rename(columns={
    'Order Number': 'OrderNumber',
    'Line Item': 'LineItem',
    'Order Date': 'OrderDate',
    'Delivery Date':'DeliveryDate',
    'Currency Code':'CurrencyCode'
},inplace=True)
```

```python
sales.shape
```

```python
sales.info()
```

```python
#Converting Data types

sales['OrderDate']= pd.to_datetime(sales['OrderDate'])
sales['DeliveryDate']= pd.to_datetime(sales['DeliveryDate'])
```

```python
# Droping delivery date row as it has 70% percent of NaN values

sales.drop('DeliveryDate', axis=1, inplace=True)
```

```python
#saving cleaned data
sales.to_csv('Sales_cleaned.csv', index=False)
```
